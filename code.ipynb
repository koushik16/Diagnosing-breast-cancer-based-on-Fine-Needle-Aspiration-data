```python
# Finding the correlation between the input features and the target feature(diagnosis)
corr_matrix = data.corr()
corr_matrix
```

```python
# The input features that show greater correlation with the target, used to train the model
prediction_vars = ['radius_mean', 'perimeter_mean', 'area_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean', 'radius_se', 'area_se', 
                   'radius_worst', 'perimeter_worst', 'area_worst', 'compactness_worst', 'concavity_worst', 'concave points_worst', 'diagnosis']
```

```python
# Splitting the data into training and testing data
train, test= train_test_split(data,test_size=0.3,random_state=42)
```

```python
train_x = train[prediction_vars]
train_y = train['diagnosis']
test_x = test[prediction_vars]
test_y = test['diagnosis']
```

```python
# dataset = data.to_numpy()
train_x = train_x.to_numpy()
train_y = train_y.to_numpy()
test_x = test_x.to_numpy()
test_y = test_y.to_numpy()
```

```python
train_x.shape
```

```python
# Calculating the Euclidean distance
def find_dist(bunch,center):
  distances = np.zeros((len(bunch),2))
  for i in range(2):
    distances[:,i] = ((bunch-center[i])**2).sum(axis=1)**0.5
  return distances
```

```python
#  K-Means
def function_for_clustering(train_x,train_y):
  seeds = [4,7,11,34,40,67,71,80]
  for i in seeds:
    np.random.seed(i)

    #finding the centroids
    centers = train_x[np.random.choice(len(train_x), 2, replace=False)]
    closest = np.zeros(len(train_x)).astype(int)

    while True:
        old_closest = closest.copy()
        distance = find_dist(train_x,centers)
        closest = np.argmin(distance, axis=1)

        for i in range(2):
          centers[i, :] = train_x[closest == i].mean(axis=0)
        
        if all(closest == old_closest):
          break
    print(accuracy_score(train_y,closest))


function_for_clustering(train_x,train_y)
```

**After running the k-means model multiple times, we observe that the accuracy is almost constant. The reason is that the centroids ultimately converge, considering that the points are uniformly distributed over a part in a two-cluster problem like this one.**

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(train_x, train_y)
print(knn.score(test_x,test_y))
```

**Comparing the results from K-Means and KNN, using a supervised model(KNN) results in a better accuracy**
